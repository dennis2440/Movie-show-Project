```python
import pandas as pd
#reading the 1st ten characters of dataframe df1 
df1 = pd.read_csv('imdb_movies_shows.csv')
df1.head(10)
                  
```

```python
#reading the 1st ten characters of dataframe df2
# and setting the index column to be 1st column in dataframe
df2 = pd.read_csv('Netflix TV Shows and Movies.csv')
df2.set_index('index',inplace=True)
df2.head(10)
```

```python
# Merge the two DataFrames on the 'imdb_id' column
merged_data = pd.merge(df2, df1, on='imdb_id')

# Define the columns you want to drop
columns_to_drop = ['title_y' ,'type_y','release_year_y','age_certification_y', 'runtime_y','imdb_score_y','imdb_votes_y','id']

# Drop the specified columns from the merged DataFrame
final_data = merged_data.drop(columns_to_drop, axis=1)
final_data

# Rename the columns
rename_col = final_data.rename(columns={'title_x': 'Title','type_x': 'Type','description': 'Description','release_year_x': 'Release Year','age_certification_x': 'Age Certification','runtime_x': 'Runtime','imdb_id': 'IMDB ID','imdb_score_x': 'IMDB Score','imdb_votes_x': 'IMDB Votes','genres': 'Genres','production_countries': 'Production Countries','seasons': 'Seasons'})
rename_col
```

```python
# moving the genre col to the 3rd index
def rearrange_columns(rename_col):
    cols = rename_col.columns.tolist()
    genre_index = cols.index('Genres')
    cols.insert(2, cols.pop(genre_index))
    return rename_col[cols]

# assign the changes to df3
df3 = rearrange_columns(rename_col)
df3.head(10)
```

```python
# checking the nan values in seasons column
rows_with_nan = df3[df3['Seasons'].isnull()]

# print rows_with_nan
# most movie rows have nan values they dont have seasons.
print(rows_with_nan)
```

```python
#replacing the nan values on the seasons column with 0
df3['Seasons'].fillna(0, inplace=True)
df3.head(10)
```

```python
#sort the dataframe according to movie/show relaese year
df4 = df3.sort_values('Release Year')

# reset the indexes of the dataset in a new index in assending order.
df5 = df4.reset_index(drop = 'True')
df5.head(10)

```

```python
# engineering a new feature 'rating' based on the imdb_rating
# Add a new 'rating' column based on the imdb_rating
df5['Rating'] = pd.cut(df5['IMDB Score'],
                      bins=[0, 5, 7, 10],
                      labels=['Poor', 'Average', 'Best'])

# Display the new DataFrame
print(df5)
```

```python
# converting df5 to a dataframe df6
df6 = pd.DataFrame(df5)

# Define  new path where to save the dataset
new_dataset_path = 'Movie_shows(update).csv'

# Save the final DataFrame as a new CSV
df6.to_csv(new_dataset_path, index=False)
```

```python
data = pd.read_csv('Movie_shows(update).csv')
data.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

# Set seaborn theme
sns.set_theme(style="whitegrid")

# Create bar plot with seaborn
movie_types = data['Type'].value_counts()
sns.barplot(x=movie_types.index, y=movie_types, palette="muted")

# Set labels and title
plt.xlabel('Type')
plt.ylabel('Frequency')
plt.title('Frequency between Movies and Shows')

# Add gridlines 
plt.grid(True)

# Show the plot
plt.show()

plt.savefig('Frequency_between_Movies_and_Shows')
```

```python
#pie chart
movie_types = data['Type'].value_counts()
plt.pie(movie_types, labels=movie_types.index, autopct='%1.1f%%')
plt.title('Proportion of Each Movie Type')
plt.show()
```

```python
# Set seaborn theme
sns.set_theme(style="whitegrid")

# scatter plot with seaborn
sns.scatterplot(x=data['IMDB Score'], y=data['Release Year'] )

# Set labels and title
plt.xlabel('IMDB Score')
plt.ylabel('Release Year')
plt.title('IMDB Score vs Release Year')

# Show the plot
plt.show()
```

```python
# histogram showing the frequency of imdb score's
plt.hist(data['IMDB Score'], bins=20)
plt.xlabel('IMDB Score')
plt.ylabel('Frequency')
plt.title('Distribution of IMDB Score among Movies & Shows')
plt.show()
```

```python
#regression plot
# get the first 300 rows
data_300 = data.head(500)

# run code on the limited data
sns.regplot(x="IMDB Score", y="IMDB Votes", data=data_300)
plt.xlabel('IMDB Score')
plt.ylabel('IMDB Votes')
plt.title('Regression of IMDB Score and Votes')

# Add ytics code to the y-axis
ytick_values = [700000, 600000, 500000, 400000, 300000, 200000, 100000, 0]
plt.yticks(ytick_values)
plt.show()
```
